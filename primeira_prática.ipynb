### criando um dicionário

> dicionário em Python que associa um número (chave) a um nome de mês (valor).O dicionário permite uma associação rápida e flexível entre as informações, e pode ser muito útil em diversas situações em que precisamos mapear valores para outras informações. Atraves disso permite a associação de informações de forma mais flexível e dinâmica. Cada chave é um número inteiro que representa um mês e cada valor é uma string que representa o nome do mês correspondente.

```python
meses = {1: 'Janeiro', 2: 'Fevereiro', 3: 'Março'}
```

### mapear o número mês para string mês

> a função dt.month é um método do pandas que é aplicado a um objeto do tipo datetime e retorna uma série com o mês correspondente a cada data. A função map é utilizada em seguida para mapear o número do mês para o seu respectivo nome.Ela também aplica a função especificada a cada elemento da série, ou seja, a cada número de mês extraído da coluna 'Data' após a conversão para um objeto de data e hora (datetime). Neste caso, a função especificada é meses, que é o dicionário que mapeia cada número de mês para seu nome correspondente. A linha de código df['Data']=pd.to_datetime(df['Data']).dt.month.map(meses) cria uma nova coluna chamada 'Data' no DataFrame df, que contém os nomes dos meses correspondentes aos valores da coluna 'Data' original.

```python
df['Data'] = pd.to_datetime(df['Data']).dt.month.map(meses)
```

```python
df
```

> é um método que pode ser aplicado em um objeto do tipo de datetime da biblioteca pandas usado para transformar datetime em string, selecionado apenas o horário. também facilita a leitura.

```python
df['Horário'] = df['Horário'].dt.strftime('%H')
```

```python
df
```

> inicia-se com lista vazia para ir preenchendo os espaços com o script;

> mapeando e transformando instring para inteiro;

> comando Append vai inserir Manhã quando o critério for verdadeiro na lista time;

> substitui a coluna horário pela lista time;

> utiliza os horários numéricos para definir manhã, tarde e noite.

```python
time = []
for i in list(map(int,df['Horário'])):
  if i > 8 and i < 13:
    time.append('Manhã')
  elif i > 12 and i < 19:
      time.append('Tarde')
  else:
        time.append('Noite')
df['Horário'] = time
```

# qual a linha de produtos tem melhor avaliação?

### agrupando dados nesse caso a coluna linha de produto com Avaliação, tirando a média das avaliações, e o comando reset_index(drop=False) evita apagar o indíce dos dados



> linha de produto é passado como uma string entre parênteses indicando que o nome da coluna que será usado para agrupar os dados;

> avaliação é passado entre colchetes indicando que a coluna cujos valores serão agregados dentro de cada grupo criado pela função groupby.

```python
Agrupamento = df.groupby('Linha de produto')['Avaliação'].mean().reset_index(drop=False)
```

```python
Agrupamento
```

### apenas arredondando para uma casa decimal

```python
Agrupando_avaliação = Agrupamento.round(1)
```

```python
Agrupando_avaliação
```

> está como false pois os valores estão sendo classificados em ordem decrescente.


```python
Agrupando_avaliação = Agrupando_avaliação.sort_values(by = 'Avaliação', ascending = False)
```

### selecionando as cores do gráfico em seguida as revertendo

```python
Figura = sns.color_palette('gnuplot')
Figura.reverse()
```

## **ax é a variável que indica o parâmetro para o matplolib plotar o gráfico, no caso um gráfico de barras barplot, em que o eixo y é a linha de produto e o eixo x a média da avaliação**

> foi feito o gráfico utilizando a biblioteca seaborn. Foram usados os dados de Linha de produto e Avaliação, para ter uma resposta de quais produtos foram melhor avaliados.

```python
ax = sns.barplot(y = 'Linha de produto', x = 'Avaliação', data = Agrupando_avaliação, palette = Figura)
```
